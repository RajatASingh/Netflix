<h1 style = "font-family:serif;"> Deal with null values</h1>

Missing data is disappointing because it feels like a missed opportunity. If the evidence is there to support it, you can salvage null values by transforming them into zeros or another value based on the other values in the column. But in most cases, it will be impossible to tell what the value should have been and you will need to remove the entire row. With large datasets, you should expect to remove rows with missing data so you can take advantage of the predictive power of each feature.

<h1 style = "font-family:serif;">Task 6: </h1>

<h1 style = "font-family:serif;"> Deal with nested Data </h1>

Nested values in a column can create problems to get accurate results. To avoid this we need to split.


<h2 style = "font-family:serif;">Working with nested "Cast" columns </h2>


```python
split_and_stacked_cast = pd.DataFrame(df["cast"].apply(lambda x : str(x).split(", ")).tolist() , 
                             index= df["title"]
                            ).stack().to_frame().reset_index().drop(columns= "level_1").rename(columns = {0 :"actors"})
```

```python
split_and_stacked_cast
```


<h2 style = "font-family:serif;">Working with nested "Director" column </h2>


```python
split_and_stacked_director = pd.DataFrame(df["director"].apply(lambda x : str(x).split(", ")).tolist() , 
                             index= df["title"]
                            ).stack().to_frame().reset_index().drop(columns= "level_1").rename(columns = {0 : "directors"})
```

```python
split_and_stacked_director.head(10)
```

<h2 style = "font-family:serif;">Working with nested "Country" column </h2>


```python
split_and_stacked_country = pd.DataFrame(df["country"].apply(lambda x : str(x).split(", ")).tolist() , 
                             index= df["title"]
                            ).stack().to_frame().reset_index().drop(columns= "level_1").rename(columns = {0 : "countries"})
```

```python
split_and_stacked_country.head(10)
```


<h2 style = "font-family:serif;">Working with nested "Listed In" column </h2>


```python
split_and_stacked_listedIn = pd.DataFrame(df["listed_in"].apply(lambda x : str(x).split(", ")).tolist() , 
                             index= df["title"]
                            ).stack().to_frame().reset_index().drop(columns= "level_1").rename(columns = {0 : "listed_in_"})
```

```python
split_and_stacked_listedIn.head(10)
```


<h2 style = "font-family:serif;"> Merging ( split_and_stacked_country , split_and_stacked_director , split_and_stacked_acst, split_and_stacked_listedIn)  in the real DataFrame</h2>

```python
df_new = df.merge(split_and_stacked_cast ,
         on= "title").merge(split_and_stacked_country ,
         on="title").merge(split_and_stacked_director , on= "title"
                          ).merge(split_and_stacked_listedIn, on= "title"
                                 ).drop(columns= ["cast", "country" , "director", "listed_in"])
```

```python
df_new.head(10)
```

```python
df_new = df_new.replace("nan" , np.nan)
```

```python
# Changing position of some columns:

df_new = df_new.loc[: ,[ 'type', 'title' , 'directors', 'actors', 'countries' ,'date_added', 'release_year', 'rating', 'duration','listed_in_']]
               
```

```python
# Modifying name of some columns

df_new = df_new.rename(columns = {"directors" : "director" , "countries" : "country"} )
```
